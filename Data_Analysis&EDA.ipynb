```python
education_counts = df['education'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Education')
plt.show()
```

```python
loan_count = df['loan'].value_counts()
deposit_count = df['deposit'].value_counts()
c = np.concatenate((loan_count.values, deposit_count.values))
labels = list(loan_count.index) + list(deposit_count.index)
plt.figure(figsize=(8, 8))
plt.pie(c, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Loan and Deposit')
plt.show()
```

```python
# Pie chart for 'age'
age_counts = df['age'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(age_counts, labels=age_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Age')
plt.show()

# Pie chart for 'month'
month_counts = df['month'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(month_counts, labels=month_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Month')
plt.show()

# Pie chart for 'duration'
duration_counts = df['duration'].value_counts()
plt.figure(figsize=(6, 6))
# Limit the number of slices for duration to the top 10 for better visualization
top_duration_counts = duration_counts.head(30)
plt.pie(top_duration_counts, labels=top_duration_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Duration (Top 10)')
plt.show()
```

```python
# Convert specific object type columns to numerical representations
for col in ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'deposit']:
    df[col] = pd.factorize(df[col])[0]
numerical_features = df.select_dtypes(include=np.number).columns
plt.figure(figsize=(10, 5))
correlation_matrix = df[numerical_features].corr()
sns.heatmap(correlation_matrix, cmap="BrBG", annot=True)
plt.show()
```
